# Modify Datas in Pandas

```python
import pandas as pd

data = {
    "Name":["Rajendra Chimala","Binod Bohara","Khum Chy","Reshma Bohara","Hari Chy","Sumit Parki"],
    "Age":[10,20,30,20,25,21],
    "Salary":[30000,40000,35000,45000,55000,64000],
    "PS":[87,34,50,90,85,99],
    "City":["Doti","Achham","Kailali","Doti","Kailali","Kailali"]
}

df = pd.DataFrame(data)

print(df)

df["Bonus"] = df['Salary'] *0.1

ID = ["GMC-001","GMC-002","GMC-003","GMC-004","GMC-005","GMC-006"]
df.insert(0,"Employee ID",ID)

print(df)


# Removing the Columns

# df.drop(columns=["ColumnNeme"], inplace=True)

df.drop(columns=["City"],inplace=True)
df
```

# Handling Missing Datas

```python
data = {
    "Name":["Rajendra Chimala","Binod Bohara",None,"Reshma Bohara","Hari Chy","Sumit Parki"],
    "Age":[10,20,None,20,25,21],
    "Salary":[30000,40000,None,45000,55000,64000],
    "PS":[87,34,None,90,85,99],
    "City":["Doti","Achham",None,"Doti","Kailali","Kailali"]
}

df = pd.DataFrame(data)
df
print(df.isnull())

df.dropna(axis=0,inplace=True)
df
```

```python
data = {
    "Name":["Rajendra Chimala","Binod Bohara",None,"Reshma Bohara","Hari Chy","Sumit Parki"],
    "Age":[10,20,None,20,25,21],
    "Salary":[30000,40000,None,45000,55000,64000],
    "PS":[87,34,None,90,85,99],
    "City":["Doti","Achham",None,"Doti","Kailali","Kailali"]
}

df = pd.DataFrame(data)
df


df["Name"].fillna(df["Name"].mode()[0],inplace=True)
df["City"].fillna(df["City"].mode()[0],inplace=True)
df["Age"].fillna(df["Age"].median(),inplace=True)
df["PS"].fillna(df["PS"].median(),inplace=True)
df["Salary"].fillna(df["Salary"].median(),inplace=True)


df
 
```
